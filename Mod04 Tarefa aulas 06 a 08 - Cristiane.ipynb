####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
# 1) 
#Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado ('Estado'ou Município), 
#e que devolve os dados da consulta no formato DataFrame.

def consultaAPI(url, uf):
    r         = requests.get(f"{url}?uf={uf.upper()}&tipo_interessado=Estado")
    data_json = r.json()
    df        = pd.DataFrame(data_json['items'])
    return df
````

````python
consultaAPI('http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl','pr')
````

````python
# 2) 
#Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com status em 'Arquivado por decurso de prazo' 
#estão registradas?

resultAPI = consultaAPI('http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl','mg')
#resultAPI[resultAPI['status'] == 'Arquivado por decurso de prazo']

resultAPI['id_pleito'][resultAPI['status'] == 'Arquivado por decurso de prazo'].count()
````

````python
# 3) Qual é o município da Bahia com mais solicitações deferidas?
resultAPI = consultaAPI('http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl','ba')
resultAPI['id_pleito'][resultAPI['status'] == 'Deferido'].value_counts()
# Não achei a coluna município
````

````python
# 4) Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'
resultAPI[resultAPI['tipo_interessado'] == 'Estado'].to_csv('./dataframe_solictacoes_ba.csv', index= False)
````
