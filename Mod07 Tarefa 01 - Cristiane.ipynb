    Dica: 
        1. inicie uma lista vazia

        3. conte o número de categorias dessa variável 
        4. acumule essa informação de 3. na lista que você criou em 1.

````python
# 1)
df = pd.DataFrame({
    'Nome': base.columns,
    'Tipo': base.dtypes
})
df
````

````python
# 2)
base2 = base.nunique()

lista = []
for n in base2:
    lista.append(n)

df['qtd_categorias'] = lista
df
````

````python
# 3) 
df2 = pd.get_dummies(base, columns=['sexo','posse_de_veiculo','posse_de_imovel', 'tipo_renda',
                                'educacao','estado_civil'], drop_first=True)
df2.head()
````
